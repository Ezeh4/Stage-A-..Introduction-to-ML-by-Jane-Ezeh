```python
# import necessary libraries
import pandas as pd
import numpy as np
```

```python
# download dataframe
url = 'https://github.com/WalePhenomenon/climate_change/blob/master/fuel_ferc1.csv?raw=true'
fuel_data = pd.read_csv(url, error_bad_lines=False)
fuel_data.describe(include='all')
```

 

## Data spliting tutorial

```python
# Split fuel_data
fuel_df1 = fuel_data.iloc[0:19000].reset_index(drop=True)
fuel_df2 = fuel_data.iloc[19000:].reset_index(drop=True)
```

```python
# print fuel_df2 to check its content
fuel_df2
```

```python
# check if the total items in the table is same with the splitted dataframes fuel_df1 and fuel df2
len(fuel_data) == (len(fuel_df1) + len(fuel_df2))
```

```python
# join the original fuel data to fuel_df1 to create duplicate items
example_df = pd.concat([fuel_data, fuel_df1]).reset_index(drop=True)
```

## Data Visualization

```python
# Import library
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plot graph
plt.figure(figsize=(7,4))
plt.xticks(rotation=90)
fuel_unit = pd.DataFrame({'unit':['BBL', 'GAL', 'GRAMSU', 'KGU', 'MCF', 'MMBTU', 'MWDTH', 'MWHTH', 'TON'],
            'count':[7998, 84, 464, 110, 11354, 180, 95, 100, 8958]})
sns.barplot(data=fuel_unit, x='unit', y='count')
plt.xlabel('Fuel Unit')
```

```python
# to select part of the dataset
sample_df = fuel_data.sample(n=50, random_state=0)
len(sample_df)
```

```python
sns.regplot(sample_df['utility_id_ferc1'], y = sample_df['fuel_cost_per_mmbtu'], fit_reg=False)
```

```python
 
# Box plot
sns.boxplot(x="fuel_type_code_pudl", y="utility_id_ferc1",
            palette=["m", "g"], data=fuel_data)
```
